```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

# Datos de ejemplo
x = np.array([1, 2, 3, 4, 5, 1, 3, 3, 7, 7, 6])

# 1. Mean
def mean_manual(x):
    return np.sum(x) / len(x)

def mean_numpy(x):
    return np.mean(x)

def mean_statistics(x):
    return st.mean(x)

# Grfica para la media
plt.figure(figsize=(8, 6))
plt.hist(x, bins=range(min(x), max(x) + 1), alpha=0.7, color='blue', edgecolor='black')
plt.axvline(mean_manual(x), color='red', linestyle='dashed', linewidth=1)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mean with Manual Calculation')
plt.legend(['Mean', 'Data Distribution'])
plt.show()

# 2. Median
def median_manual(x):
    sorted_data = sorted(x)
    n = len(sorted_data)
    if n % 2 == 0:
        return (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
    else:
        return sorted_data[n // 2]

def median_numpy(x):
    return np.median(x)

def median_statistics(x):
    return st.median(x)

# Grfica para la mediana
plt.figure(figsize=(8, 6))
plt.hist(x, bins=range(min(x), max(x) + 1), alpha=0.7, color='green', edgecolor='black')
plt.axvline(median_manual(x), color='red', linestyle='dashed', linewidth=1)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Median with Manual Calculation')
plt.legend(['Median', 'Data Distribution'])
plt.show()

# 3. Mode
def mode_manual(x):
    counts = {}
    for i in x:
        counts[i] = counts.get(i, 0) + 1
    mode_value = max(counts, key=counts.get)
    return mode_value

def mode_numpy(x):
    counts = np.bincount(x)
    return np.argmax(counts)

def mode_st(x):
    return st.mode(x)

# Grfica para la moda
plt.figure(figsize=(8, 6))
plt.hist(x, bins=range(min(x), max(x) + 1), alpha=0.7, color='orange', edgecolor='black')
plt.axvline(mode_manual(x), color='red', linestyle='dashed', linewidth=1)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mode with Manual Calculation')
plt.legend(['Mode', 'Data Distribution'])
plt.show()

# 4. Range
def range_manual(x):
    return max(x) - min(x)

def range_numpy(x):
    return np.ptp(x)

# Grfica para el rango
plt.figure(figsize=(8, 6))
plt.hist(x, bins=range(min(x), max(x) + 1), alpha=0.7, color='purple', edgecolor='black')
plt.axvline(min(x), color='blue', linestyle='dashed', linewidth=1)
plt.axvline(max(x), color='blue', linestyle='dashed', linewidth=1)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range with Manual Calculation')
plt.legend(['Min', 'Max', 'Data Distribution'])
plt.show()

# 5. Variance
def variance_manual(x):
    mean = mean_manual(x)
    variance = 0
    for i in x:
        variance += (mean - i) ** 2
    return variance / (len(x) - 1)

def variance_numpy(x):
    return np.var(x)

def variance_st(x):
    return st.variance(x)

# Grfica para la varianza
plt.figure(figsize=(8, 6))
plt.hist(x, bins=range(min(x), max(x) + 1), alpha=0.7, color='cyan', edgecolor='black')
plt.axvline(mean_manual(x) - np.sqrt(variance_manual(x)), color='red', linestyle='dashed', linewidth=1)
plt.axvline(mean_manual(x) + np.sqrt(variance_manual(x)), color='red', linestyle='dashed', linewidth=1)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Variance with Manual Calculation')
plt.legend(['Mean - Std Dev', 'Mean + Std Dev', 'Data Distribution'])
plt.show()

# Standar deviation
def standard_deviation_manual(x):
    return np.sqrt(variance_manual(x))

def standard_deviation_numpy(x):
    return np.std(x)

def standard_deviation_st(x):
    return st.stdev(x)

# Grfica para la desviacin estndar
plt.figure(figsize=(8, 6))
plt.hist(x, bins=range(min(x), max(x) + 1), alpha=0.7, color='magenta', edgecolor='black')
plt.axvline(mean_manual(x) - np.sqrt(variance_manual(x)), color='red', linestyle='dashed', linewidth=1)
plt.axvline(mean_manual(x) + np.sqrt(variance_manual(x)), color='red', linestyle='dashed', linewidth=1)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Standard Deviation with Manual Calculation')
plt.legend(['Mean - Std Dev', 'Mean + Std Dev', 'Data Distribution'])
plt.show()
```
